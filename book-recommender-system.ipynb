## Popularity Based Recommender System

```python
#merge rating and books
ratings_with_name=ratings.merge(books,on="ISBN")
ratings.merge(books,on="ISBN").shape
```

```python
ratings_with_name
```

```python
#first count how many votes for each book
ratings_with_name.groupby('Book-Title').count()
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating']
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()

num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()

avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=avg_rating_df.merge(num_rating_df,on="Book-Title")
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
#now we need publisher name,image of books etc

popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title') 
popular_df
```

```python
popular_df=popular_df[["Book-Title","Book-Author","Image-URL-S","num_ratings","avg_ratings","Image-URL-M"]]
```

```python
x=ratings_with_name.groupby("User-ID").count()['Book-Rating']>200
users_above_200_books=x[x].index #All true indexed
users_above_200_books
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(users_above_200_books)] #based on user
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
final_ratings
 
```

```python
pt=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
#706 books and  810 users
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_scores=cosine_similarity(pt) 
```

```python
 similarity_scores
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    distances=similarity_scores[index]
    similar_items=sorted(list(enumerate(distances)),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')
        ['Book-Title']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')
        ['Book-Author']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')
        ['Image-URL-M']))
        data.append(item)
    return data
    
```

```python
 recommend("Message in a Bottle")
```
